```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
from langchain.embeddings.base import Embeddings

class MySentenceTransformerEmbeddings(Embeddings):
    def __init__(self, model):
        self.model = model

    def embed_query(self, text):
        return self.model.encode(text).tolist()

    def embed_documents(self, texts):
        return [self.model.encode(text).tolist() for text in texts]
```

```python
from langchain.vectorstores import Chroma
```

```python
embedding = MySentenceTransformerEmbeddings(model)
```

```python
!pip install chromadb
```

```python
persist_directory = 'docs/chroma'
vectordb = Chroma.from_documents(documents =docs,embedding=embedding, persist_directory=persist_directory )
```

```python
vectordb._collection.count()
```

```python
question = "What is Encoder part in Rt Mono ?"
docans = vectordb.similarity_search(question,k=3)
```

```python
docans[0].page_content
```

```python
from transformers import pipeline
from langchain.llms import HuggingFacePipeline

# Create text2text-generation pipeline
flan_pipeline = pipeline("text2text-generation", model="google/flan-t5-base", max_length=512)

# Wrap inside a LangChain LLM
llm = HuggingFacePipeline(pipeline=flan_pipeline)
```

```python
query = "What si feedforward neural network"

docs = vectordb.similarity_search(query, k=3)
```

```python
# Combine retrieved docs
context = "\n\n".join([doc.page_content for doc in docs])

prompt = f"""Use the following context to answer the question:

{context}

Question: {query}

Answer:"""

# Run through LLM
response = llm(prompt)

print("Answer:", response)
```
