```python
import nltk
import pandas as pd
```

```python
data = []
with open('MT Eng Tam Dataset.txt', 'r', encoding='utf-8') as f:
    for line in f:
        data.append(line)  # strip() removes leading/trailing whitespace, including newlines

source = []
target = []
for line in data:
    line = line.split("        ")
    source.append(line[0])
    target.append(line[1])
print(source)
```

```python
print(target)
```

```python
df = pd.DataFrame({"English":source,"Tamil":target})
df.head()
```

```python
source_vocab = set(''.join(source))
target_vocab = set(''.join(target))
print(source_vocab)
```

```python
print(target_vocab)
```

```python
source_vocab_len = len(source_vocab)
target_vocab_len = len(target_vocab)
```

```python
print(source_vocab_len)
print(target_vocab_len)
```

```python
source_char_int = {char:int for int,char in enumerate(source_vocab)}
target_char_int = {char:int for int,char in enumerate(target_vocab)}
source_int_char = {int: char for char,int in source_char_int.items()}
target_int_char = {int: char for char,int in target_char_int.items()}

print(source_char_int)
print(target_char_int)
```

```python
print(target_int_char)
```

```python
source_sequence = []
target_sequence = []
for text in source:
    seq = [source_char_int[char] for char in text]
    source_sequence.append(seq)
for text in target:
    seq = [target_char_int[char] for char in text]
    target_sequence.append(seq)
```

```python
print(source_sequence)
print(target_sequence)
```

```python
max_seq_length = max(max(len(seq) for seq in source_sequence),max(len(seq) for seq in target_sequence))
max_seq_length
import tensorflow as tf
```

```python
source_sequence = tf.keras.preprocessing.sequence.pad_sequences(source_sequence,max_seq_length,padding='post')
target_sequence = tf.keras.preprocessing.sequence.pad_sequences(target_sequence,max_seq_length,padding='post')
```

```python
print(source_sequence)
```

```python
print(target_sequence)
```

```python
import numpy as np
model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(source_vocab_len,64,input_length=max_seq_length),
    tf.keras.layers.SimpleRNN(128,return_sequences=True),
    tf.keras.layers.Dense(target_vocab_len,activation='softmax')
])
model.compile(optimizer='adam',loss= 'categorical_crossentropy',metrics=['accuracy'])
```

```python
target_seq_onehot = np.array([tf.keras.utils.to_categorical(seq,target_vocab_len) for seq in target_sequence])
```

```python
model.fit(
    source_sequence,
    target_seq_onehot,
    epochs = 1000,
    batch_size = 4,
    verbose = 1
)
```

```python
def translate(text):
    input_seq = [source_char_int[char] for char in text if char in source_char_int]
    input_seq = tf.keras.preprocessing.sequence.pad_sequences([input_seq],maxlen = max_seq_length,padding='post')
    output_sequence = model.predict(input_seq)[0]
    output = ''
    for char_prob in output_sequence:
        char_idx = np.argmax(char_prob)
        char = target_int_char[char_idx]
        output += char
    return output
```

```python
s = "Hi how are you"
print(translate(s))
```
