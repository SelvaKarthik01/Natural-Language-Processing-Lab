```python
import nltk
import pandas as pd
import numpy as np
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
df = pd.read_excel('SASTRA University.xlsx');
df.head()
```

```python
df.columns = ['sentences']
df.head()
```

```python
def preprocess_text(sentence):
    tokens = word_tokenize(sentence.lower())
    tokens = [token for token in tokens if token.isalpha()]
    return tokens
df['tokens'] = df['sentences'].apply(preprocess_text)
```

```python
def bi_gram(tokens):
    return list(ngrams(tokens,2))
df['bigrams'] = df['tokens'].apply(bi_gram)
df.head()
```

```python
all_words = [word for words in df['tokens'] for word in words]
word_freq = Counter(all_words)
N = sum(word_freq.values())
```

```python
all_bigrams = [bigram for bigrams in df['bigrams'] for bigram in bigrams]
bigram_freq = Counter(all_bigrams)
```

```python
print(word_freq)
print(bigram_freq)
```

```python
import math
critical_value = 1.96
print('t-test for all the bigrams : ')
for bigram in all_bigrams:
    a,b = bigram
    p_ab=bigram_freq.get(bigram,0)
    p_a = word_freq.get(a,0)
    p_b = word_freq.get(b,0)
    if p_a == 0 or p_b == 0 or p_ab == 0:
        print('Skipping Bigram due to low count in dataset')
    else:
        t_value =((p_ab/N)-(p_a*p_b/(N*N)))/np.sqrt(p_ab/(N*N))
        if t_value > critical_value:
            print(bigram,"t_value",t_value,"Strong Collocation")
        else:
            print(bigram,"t_value",t_value,"Weak Collocation")
```
